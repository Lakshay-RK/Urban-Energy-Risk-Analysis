```python
import numpy as np
import pandas as pd
import libpysal
from esda.moran import Moran
from spreg import ML_Lag 
from libpysal.weights import lat2W

grid_dim = 10 
n = grid_dim**2

x, y = np.indices((grid_dim, grid_dim))
dist_from_center = np.sqrt((x - 5)**2 + (y - 5)**2)

income = 50000 + (dist_from_center * 2000) + np.random.normal(0, 2000, n).reshape(10,10)
building_age = 50 - (dist_from_center * 3) + np.random.normal(0, 5, n).reshape(10,10)

energy_risk = (building_age.flatten() * 0.5) - (income.flatten() * 0.0005) + 30
energy_risk += np.random.normal(0, 1, n) 

w = lat2W(grid_dim, grid_dim)
w.transform = 'R' 

mi = Moran(energy_risk, w)
print(f"--- Spatial Analysis ---")
print(f"Clustering Statistic (Moran's I): {mi.I:.4f}")
print(f"P-value: {mi.p_sim:.4f} (Significant if < 0.05)\n")

X = np.column_stack([income.flatten(), building_age.flatten()])
y = energy_risk.reshape(n, 1)

model = ML_Lag(y, X, w=w, name_x=['Income', 'Bld_Age'], name_y='Energy_Risk')

print(f"--- Regression Results ---")
print(model.summary)
```
